```python
import pandas as pd
import numpy as np
ball_by_ball = pd.read_csv('../resources/train/ball_by_ball_data.csv')
ball_by_ball.head(n=2)
```

```python
bat_vs_bowl = ball_by_ball.groupby(['match_id', 'inning', 'batsman_id', 'bowler_id']).agg({
    'batsman_runs': np.sum, 'ball':np.size}).reset_index()
bat_vs_bowl.head(n=2)
```

```python
matches = [pd.read_csv('../resources/train/match_data.csv')[['match_id', 'team1_id', 'team2_id']], (
    pd.read_csv('../resources/test/matches_2018.csv')[['match_id', 'team1_id', 'team2_id']])]
matches = pd.concat(matches)
matches['Season'] = matches.apply(lambda x: int(x.match_id[:4]), axis = 1)
matches.head(n=2)
```

```python
rosters = pd.read_csv('../resources/train/player_rosters.csv')
rosters.head(n=2)
```

```python
pd.concat([rosters.merge(matches, left_on=['Season', 'Team'], right_on=['Season', 'team2_id'], how='left'),
          rosters.merge(matches, left_on=['Season', 'Team'], right_on=['Season', 'team1_id'], how='left')])
```
