```python
import pandas as pd
ball_by_ball = pd.read_csv('../resources/train/ball_by_ball_data.csv')
ball_by_ball = ball_by_ball[ball_by_ball['is_super_over'] == 0]
key_teams = pd.read_csv('../resources/train/key_teams.csv')
match_data = pd.read_csv('../resources/train/match_data.csv')
player_roasters = pd.read_csv('../resources/train/player_rosters.csv')
```

```python
ball_by_ball.head(n=2)
```

```python
#per over stats
import numpy as np

bat_vs_bowl = ball_by_ball.groupby(['match_id', 'inning', 'batting_team', 'bowling_team', 'batsman_id', 
                                     'bowler_id']).agg({'batsman_runs': np.sum})

bat_vs_bowl = bat_vs_bowl.reset_index()
bat_vs_bowl.head(n=5)
```

```python
#unique wasy to get out
dism_types = ball_by_ball.dismissal_kind.dropna(how='all').unique()
#credit to bowler
dis_by_bowl = ['caught', 'bowled', 'lbw', 'caught and bowled', 'stumped', 'hit wicket']
dis_other = ['run out', 'retired hurt', 'obstructing the field']

wicket_del = ball_by_ball[ball_by_ball['dismissal_kind'].isin(list(dism_types))]

wicket_del = wicket_del[['match_id', 'player_dismissed_id', 'bowler_id', 'dismissal_kind']]
wicket_del.head(n=2)
```

```python
match_stat = pd.merge(bat_vs_bowl, wicket_del, how='left', 
         left_on=['match_id', 'batsman_id','bowler_id'], right_on=['match_id', 'player_dismissed_id','bowler_id',])
match_stat = match_stat[['match_id', 'inning', 'batting_team', 'bowling_team', 'batsman_id', 'bowler_id', 'batsman_runs', 'dismissal_kind']]
match_stat = match_stat.merge(match_data[['match_id', 'venue_id']], how='left', left_on=['match_id'], right_on=['match_id'])
match_stat.head(n=2)
```
